```python
def plotDist(a):
    
    f, ax = plt.subplots(2, 2, figsize=(8,8))
    ax[0, 0].imshow(a, cmap = 'Greys', interpolation = 'None')
    ax[0, 0].set_title('Adjacency Matrix')
    
    D = np.corrcoef(a)
    ax[1, 0].imshow(D, cmap = 'Greys', interpolation = 'None')
    ax[1, 0].set_title('Correlation coeff.')
    
    dVec = spt.distance.pdist(a, metric = 'euclidean')
    D = spt.distance.squareform(dVec)
    ax[0, 1].imshow(D, cmap = 'Greys', interpolation = 'None')
    ax[0, 1].set_title('Euclidean Dist.')
    
    dVec = spt.distance.pdist(a, metric = 'cosine')
    D = spt.distance.squareform(dVec)
    ax[1, 1].imshow(D, cmap = 'Greys', interpolation = 'None')
    ax[1, 1].set_title('Cosine Dist.')
plt.savefig("g:gMatrix.png")
plotDist(a)
```

```python
def get_province(s):
    s = s.replace('  ',' ').replace('-',' ').replace(',',' ').strip();
    
    for p in provinces:
        if(s.__contains__(p)):
            return p;
        elif(s.__contains__('استان ' + p)):
            return p;

def get_name_details(name):
    name = name.strip();
    
    namedetails = {'name':name,'faname':'','enname':'','ptitle':'','percent':0,'sex':-1,'firstname':'','lastname':'','hascorporationtitle':False}
    duplicates = ['دانشگاه آزاد اسلامی', 'جهاد دانشگاهی', 'نهاد', 'پژوهشگاه', 'پژوهشکده', 'مرکز', 'دانشگاه', 'موسسه',
              'شرکت', 'پارک علم و فناوری',
              'بنیاد', 'دفتر', 'انجمن', 'وزارتخانه','وزارت']
    for d in duplicates:
        name = name.replace(d + ' ' +d,d);
        if(name.startswith(d)):
            namedetails['hascorporationtitle']=True
    
    pretitleregex = ['سرکار[ ]+خانم[ ]+دکتر', 'جناب[ ]+آقای[ ]+دکتر', 'جناب[ ]+آقای','سرکار[ ]+خانم','سازمان',
         'آقای[ ]+سید', 'خانم[ ]+سیده', 'پارک[ ]+علم[ ]+و[ ]+فناوری','دانشگاه علوم پزشکی',
         'آقای', 'دکتر', 'خانم', 'سیده', 'سید','دفتر','بنیاد','پژوهشگاه', 'پژوهشکده','نهاد','مرکز', 'دانشگاه', 'موسسه', 'شرکت']
    for ptitle in pretitleregex:
        if (re.search('^'+ptitle, name)):
            last1 = re.search('^'+ptitle, name).end();
            namedetails['ptitle'] = re.search('^'+ptitle, name).group(0).strip();
            name = name[last1:].strip();
            break;
            
    if(namedetails['ptitle']==''):
        if(name.startswith('جهاد دانشگاهی')):
            namedetails['ptitle'] = 'نهاد'
        else:
            pretitleregex =[ 'بنیاد', 'پژوهشگاه', 'پژوهشکده', 'نهاد', 'مرکز',
                     'دانشگاه', 'موسسه', 'پارک[ ]+علم[ ]+و[ ]+فناوری','شرکت']
            for ptitle in pretitleregex:
                if (re.search(ptitle, name)):
                    namedetails['ptitle'] = re.search(ptitle, name).group(0).strip();
                    break;
    
    if (name.__contains__(')')):
        name = name[0:re.search('\)', name).end() + 1]
    percent  = 0;
    name = name.replace('/','.').replace(',','.').replace('،','.').replace('¸','.')
    if (re.search('\([ ]*([۱۲۳۴۵۶۷۸۹۰]+([ ]*\.[ ]*[۱۲۳۴۵۶۷۸۹۰]+)?)[ ]*%[ ]*\)', name) != None):
        percent = float(re.search('\([ ]*([۱۲۳۴۵۶۷۸۹۰]+([ ]*\.[ ]*[۱۲۳۴۵۶۷۸۹۰]+)?)[ ]*%[ ]*\)', name).group(1).replace(' ',''))
        name = name.replace(re.search('\([ ]*([۱۲۳۴۵۶۷۸۹۰]+([ ]*\.[ ]*[۱۲۳۴۵۶۷۸۹۰]+)?)[ ]*%[ ]*\)', name).group(0), '').strip();
        
    namedetails['percent'] = percent;
    avoidedalphabets = '()-+._,.%'
    for alphabet in avoidedalphabets:
        name = name.replace(alphabet,'').strip();
    
    namedetails['name'] = name;
    namedetails['faname'] = name;
    
    reg = '[a-zA-Z][a-zA-Z -_,\(\)]+[a-zA-Z\)]';
    if(re.search(reg,name)):
        namedetails['enname']=re.search(reg,name).group(0)
        namedetails['faname']=name.replace(namedetails['enname'],'')
        
        
    realname = namedetails['faname'].replace('  ','').strip();
    if(realname.split(' ').__len__() == 2):
        namedetails['firstname'] = realname.split(' ')[0];
        namedetails['lastname'] = realname.split(' ')[1];
    else:
        pre=''
        for n in realname.split(' '):
            n = n.strip();
            pre = pre + n;
            if(n.__len__()>0):
                if(names.__contains__(pre.replace(' ',''))):
                    namedetails['firstname'] = pre;
                    namedetails['lastname'] = realname[len(pre):].strip();
                    namedetails['sex'] = names[pre.replace(' ','')];
                pre = pre + ' ';        
        
    return namedetails
        
def get_inventedOwner(s):
    listofpeopel = [];
    start = 'مالک[ ]+اختراع[ ]*:?'
    last = '(نام[ ]+مخترع|نام[ ]+مالک|موضوع[ ]+اختراع|طبقه[ ]+بندی[ ]+بین[ ]+المللی|نام[ ]+نماینده[ ]+قانونی|مالک[ ]+اختراع)[ ]*:?'

    startp = 0;
    lastp = -1;

    while (re.search(start, s[startp:])):
        name = ''
        legalpersontype = ''
        address = ''
        registerno = ''

        startp = re.search(start, s[startp:]).start() + startp
        if (re.search(last, s[startp + 1:])):
            lastp = re.search(last, s[startp + 1:]).start() + startp + 1
        else:
            lastp = len(s);

        s1 = s[startp:lastp];

        attributes = [
            {'regex': 'مالک[ ]+اختراع', 'propname': 'name'},
            {'regex': 'نوع[ ]+شخص[ ]*(حقوقی)?', 'propname': 'legalpersontype'},
            {'regex': 'شماره[ ]+ثبت', 'propname': 'registerno'},
            {'regex': 'نشانی', 'propname': 'address'},
            {'regex': 'تابعیت', 'propname': 'nationality'}
        ]

        patent = {"ptitle": '', "name": '','faname':'', "percent": 0, "legalpersontype": None, "registerno": None,
                  "address": '', 'nationality': '','enname':'','province':'','sex':-1,'firstname':'','lastname':''};

        for att in attributes:
            regex = ''
            for att1 in attributes:
                if (att1 != att):
                    regex = regex + att1['regex'] + '|';
            endregex = '(' + regex[0:-1] + ')[ ]*:?';
            startregex = att['regex'] + '[ ]*:?';

            if (re.search(startregex, s1)):
                startp1 = re.search(startregex, s1).end()
                lastp1 = len(s1) - 1;
                if (re.search(endregex, s1[startp1:])):
                    lastp1 = re.search(endregex, s1[startp1:]).start() + startp1
                patent[att['propname']] = s1[startp1:lastp1].strip()
        
        name = patent['name']
        name_details=get_name_details(name);
        patent.update(name_details);
        patent['province']=get_province(patent['address']);
        
        if(patent['ptitle'] == '' and ( patent["legalpersontype"]!= None or patent["registerno"]!= None)):
            patent['ptitle'] = '-';
        
        if(len(patent['name']) > 2 or len(patent['enname']) ):
            listofpeopel.append(patent)
        
        startp = lastp;

    return listofpeopel


def get_inventorName(s):
    listofpeopel = [];
    start = 'نام[ ]+مخترع[ ]*:?'

    last = '(مالک[ ]+اختراع|نام[ ]+مخترع|نام[ ]+مالک|موضوع[ ]+اختراع|طبقه[ ]+بندی[ ]+بین[ ]+المللی|نام[ ]+نماینده[ ]+قانونی)[ ]*:?'

    startp = 0;
    lastp = -1;

    while (re.search(start, s[startp:])):
        name = ''
        legalpersontype = ''
        address = ''
        registerno = ''

        startp = re.search(start, s[startp:]).start() + startp
        if (re.search(last, s[startp + 1:])):
            lastp = re.search(last, s[startp + 1:]).start() + startp + 1
        else:
            lastp = len(s);

        s1 = s[startp:lastp];

        attributes = [
            {'regex': 'نام[ ]+مخترع', 'propname': 'name'},
            {'regex': 'نوع[ ]+شخص[ ]*(حقوقی)?', 'propname': 'legalpersontype'},
            {'regex': 'شماره[ ]+ثبت', 'propname': 'registerno'},
            {'regex': 'نشانی', 'propname': 'address'},
            {'regex': 'تابعیت', 'propname': 'nationality'}
        ]

        patent = {"ptitle": '', "name": '', "percent": 0, "legalpersontype": None, "registerno": None,
                  "address": '', 'nationality': '','enname':'','province':'','sex':-1,'firstname':'','lastname':''};

        for att in attributes:
            regex = ''
            for att1 in attributes:
                if (att1 != att):
                    regex = regex + att1['regex'] + '|';
            endregex = '(' + regex[0:-1] + ')[ ]*:?';
            startregex = att['regex'] + '[ ]*:?';

            if (re.search(startregex, s1)):
                startp1 = re.search(startregex, s1).end()
                lastp1 = len(s1) - 1;
                if (re.search(endregex, s1[startp1:])):
                    lastp1 = re.search(endregex, s1[startp1:]).start() + startp1
                patent[att['propname']] = s1[startp1:lastp1].strip()
        
        
        name = patent['name'].strip()
        name_details=get_name_details(name);
        patent.update(name_details);
        patent['province']=get_province(patent['address']);
        
        if(patent['ptitle'] == '' and ( patent["legalpersontype"]!= None or patent["registerno"]!= None)):
            patent['ptitle'] = '-';
        
        if(len(patent['name']) > 2 or len(patent['enname']) ):
            listofpeopel.append(patent)
            
        startp = lastp;

    return listofpeopel


def get_inventedSubject(s):
    start = 'موضوع[ ]+اختراع[ ]*:?'
    end = 'طبقه[ ]+بندی[ ]+بین[ ]+المللی[ ]*:?'

    a = re.findall(r'{}(.*?){}'.format(start, end), s)
    if (a != None and len(a) > 0 and len(a[0]) > 2):
        return a
    return re.findall(r'{}(.*?)$'.format(start), s)


def get_internationalCategory(s):
    start = 'طبقه[ ]+بندی[ ]+بین[ ]+المللی[ ]*:?'
    last = '(تاریخ|کشور|مدت)[ ]*:?'
    end = 'مدت'
    
    startp = 0;
    lastp = -1;

    classification = {"IPC": '', "country": '', "releasedate": '','IPC3':None};
    name = ''
    legalpersontype = ''
    address = ''
    registerno = ''
    if(re.search(start, s[startp:]) == None):
        return classification;
    
    startp = re.search(start, s[startp:]).start() + startp
    if (re.search(last, s[startp + 1:])):
        lastp = re.search(last, s[startp + 1:]).start() + startp + 1
    else:
        lastp = len(s);

    s1 = s[startp:lastp];

    attributes = [
        {'regex': 'طبقه[ ]+بندی[ ]+بین[ ]+المللی', 'propname': 'IPC'},
        {'regex': 'کشور', 'propname': 'country'},
        {'regex': 'تاریخ', 'propname': 'releasedate'},
    ]

    

    for att in attributes:
        regex = ''
        for att1 in attributes:
            if (att1 != att):
                regex = regex + att1['regex'] + '|';
        endregex = '(' + regex[0:-1] + ')[ ]*:?';
        startregex = att['regex'] + '[ ]*:?';

        if (re.search(startregex, s1)):
            startp1 = re.search(startregex, s1).end()
            lastp1 = len(s1) - 1;
            if (re.search(endregex, s1[startp1:])):
                lastp1 = re.search(endregex, s1[startp1:]).start() + startp1
            classification[att['propname']] = s1[startp1:lastp1].strip()
        classification['IPC'] = classification['IPC'].replace(' ','')
        ipcreg = '([A-Z][۰-۹\d]{2}[A-Z]([۰-۹\d]{1,3}/[۰-۹\d]{1,3})?\;)+';
        
        if(re.search(ipcreg,classification['IPC'])):
            classification['IPC'] = re.search(ipcreg,classification['IPC']).group(0).replace(' ','');
        
        ipcreg = '([A-Z][۰-۹\d]{2}[A-Z])';
        classification['IPC3'] = re.findall(ipcreg,classification['IPC']);
        
    return classification
    
    

```

```python
def dedup_persons(persons):
    cno = 0;
    unlabel = 0
    dic_persons = dict();
    for person in persons:
        person['cluster'] = -1;
        if(person['name'] == None ):
            unlabel = unlabel + 1
            continue;
        
        if(person['province'] == '' or person['province']==None):
            person['province'] = 'تهران'
        codedname = (person['name'].replace(' ','') +'_'+person['province']).strip();
            
        if(dic_persons.__contains__(codedname)==False):
            dic_persons[codedname] = str(cno);
            person['cluster'] = 100000+cno;
            cno = cno + 1;
        else:
            person['cluster'] = dic_persons[codedname]
        
    return persons,dic_persons,cno , unlabel   

def dedup_companies(companies):
    cno = 0;
    unlabel =0 
    dic_companies = dict();
    dic_companies1 = dict();
    for company in companies:
        precno = cno;
        company['cluster'] = -1;
        if(company['faname'] != None and company['faname'] != ''  ):
            pid = company['id']
            if(company['province'] == None or company['province'] == '' ):
                company['province'] = 'تهران'

            codedname = ('_'+company['faname']).replace(' ','').strip();

            avoidedwords = ['دولتی','ملی','کشوری','صنعتی','نهاد','فدراسیون','علوم پزشکی','صنایع','دانشگاه']
            for avoidword in avoidedwords:
                codedname = codedname.replace(avoidword,'').strip();
            codedname  =company['ptitle']+ codedname;

            if(dic_companies.__contains__(codedname)==False):
                dic_companies[codedname] = cno;
                cno = cno + 1;
                
            company['cluster'] = dic_companies[codedname] ;
            precno = dic_companies[codedname]
        
        if(company['enname']  != None and company['enname']  != ''  ):
            codedname = ('_'+company['enname']).replace(' ','').strip();

            avoidedwords = ['دولتی','ملی','کشوری','صنعتی','نهاد','فدراسیون','علوم پزشکی','صنایع','دانشگاه']
            for avoidword in avoidedwords:
                codedname = codedname.replace(avoidword,'').strip();
            codedname = company['ptitle'] + codedname;

            if(dic_companies1.__contains__(codedname)==False):
                dic_companies1[codedname] = precno;
                if(precno == cno):
                    cno = cno + 1
            
            company['cluster'] = min(precno,dic_companies1[codedname])
            
    return companies,cno,unlabel
  
```

```python
import numpy as np
def jaccard_kgram(str1,str2,k=2):
    if(str1 == None or len(str1) == 0 or str2 == None or len(str2) == 0):
        return 0
    str1 = str1.lower().replace(' ','').strip();
    str2 = str2.lower().replace(' ','').strip();
    A = dict();
    B = dict();
    for i in range(len(str1)-(k-1)):
        A[str1[i:i+k]] = 1;
    for i in range(len(str2)-(k-1)):
        B[str2[i:i+k]] = 1;
    A_and_B = 0;
    AorB = len(B);
    
    for x in A.keys():
        if(B.__contains__(x)):
            A_and_B = A_and_B + 1;
        else:
            AorB = AorB + 1
          
    return A_and_B/AorB

def sim(names,fn=jaccard_kgram):
    simmatrix = np.zeros([len(names),len(names)])
    
    for i in range(len(names)):
        for j in range(len(names)):
            value = jaccard_kgram(names[i],names[j]);
            simmatrix[i][j]= value;
            simmatrix[j][i] = value;
            
    return simmatrix;
#jaccard_kgram(' دولتی تربیت مدرس','تربیت مدرس')
#simmatrix = sim([person['enname'] for person in persons])
```
